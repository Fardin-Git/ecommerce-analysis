<h1 style="text-align:center;">Analysis</h1>


After cleaning the csv about sales data according to our wishes, we can read it in, to pursue our further analysis.

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cleaned/clean_data.csv")
df.head()
```

Now, we are able to calculate important KPIs and plot them, as a basis for our insights and recommendations for action.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
sns.set_style("whitegrid")
```

## We start with monthly revenues over the years.

```python
monthly_revenue = (
    df.groupby('YearMonth')['Revenue']
      .sum()
      .sort_index()
)

plt.figure(figsize=(8,3))
sns.lineplot(x=monthly_revenue.index.astype(str), y=monthly_revenue.values, marker="o")
plt.xticks(rotation=45)
plt.title("Revenue per month")
plt.xlabel("Month")
plt.ylabel("Revenue ()")
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

We can see that the revenue rose steadily over time until it reached its peak in November 2011 and then fell dramatically.

## The top 10 products that generated the most revenue:

```python
top_products = (
    df.groupby('Description')['Revenue']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(7,3))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 10 products by revenue")
plt.xlabel("Revenue ()")
plt.ylabel("Product")
plt.tight_layout()
plt.show()
```

We see that "Regency Cakestand 3 Tier" generates by far the most revenue, and places 2 to 4 are very close together.

## The top 10 customers who generated the most revenue:

```python
top_customers = (
    df[df['CustomerID'] != 0] #Without anonymous customers
    .groupby('CustomerID')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(7,3))
sns.barplot(x=top_customers.values, y=top_customers.index.astype(str))
plt.title("Top 10 customers by revenue")
plt.xlabel("Revenue ()")
plt.ylabel("CustomerID")
plt.tight_layout()
plt.show()
```

It is evident that places 1 and 2 generated by far the most revenue.

## The revenue generated by anonymous orders:

```python
anon_revenue = df[df['CustomerID'] == 0]['Revenue'].sum()
anon_revenue
```

We can see that the revenue by anonymous customers plays a significant role for the overall analysis.

## The top 10 countries by revenue:

```python
revenue_per_country = (
    df.groupby('Country')['Revenue']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(7,3))
sns.barplot(x=revenue_per_country.values, y=revenue_per_country.index)
plt.title("Top 10 countries by revenue")
plt.xlabel("Revenue ()")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

The United Kingdom generates by far the largest revenue, while Ireland and the Netherlands, in second and third place respectively,<br>
do not generate significant revenue but are still importat in comparison to other countries.

## The top 10 customers with the highest average revenue:

```python
avg_revenue_per_customer = (
    df[df['CustomerID'] != 0] #Without anonymous customers
      .groupby('CustomerID')['Revenue']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(7,3))
sns.barplot(x=avg_revenue_per_customer.values, y=avg_revenue_per_customer.index.astype(str))
plt.title("Top 10 Customers by average revenue")
plt.xlabel("Average Revenue ()")
plt.ylabel("CustomerID")

plt.tight_layout()
plt.show()
```

These are customers who not only place large individual orders but also generate high revenue across multiple orders. <br>Customer 16000 is the clear number 1.

## The number of orders per month:

```python
orders_per_month = (
    df.groupby('YearMonth')['InvoiceNo']
      .nunique()  
      .sort_index()
)

plt.figure(figsize=(7,3))
sns.lineplot(x=orders_per_month.index.astype(str), y=orders_per_month.values, marker="o")
plt.xticks(rotation=45)
plt.title("Number of orders per month")
plt.xlabel("Month")
plt.ylabel("Number of orders")
plt.tight_layout()
plt.show()
```

The order patterns are very similar to the sales patterns, suggesting a causal relationship.

## The average revenue per order:

```python
avg_revenue_per_order = (
    df.groupby('InvoiceNo')['Revenue']
      .sum()
      .mean()
)
avg_revenue_per_order
```

We see that the average revenue per order is very high, which suggests that companies placing large orders are also using the market.

## Recommendations for action

To avoid a repeat of the decline in orders seen from November to December 2011, several courses of action may be recommended.<br>
Marketing efforts should be present even in strong months, particularly in countries like the UK and Ireland, to target wholesalers who<br> also place large orders. <br>Furthermore, targeted offers and discounts can attract the top 10 users with the highest average revenue, especially user 1600 or user 14646, who generate<br> especially high revenue. The product range and marketing for products such as "Regency Cakestand 3 Tier", "Party Bunting" or "Jumbo bag red retrospot" should also be increased.<br> In the further course of the process, the purchasing behavior of top customers can be analyzed in order to create personalized advertising.
